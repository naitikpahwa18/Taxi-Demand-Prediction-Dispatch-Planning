# Results Summary

Loads pre-saved metrics and plots from `results/`. 
Run this notebook to review results **without re-training** the models.

```python
import pandas as pd, matplotlib.pyplot as plt, matplotlib.image as mpimg, os

csv_path = 'results/metrics.csv'
if os.path.exists(csv_path):
 metrics_df = pd.read_csv(csv_path)
 print('Model Performance:')
 display(metrics_df)
else:
 print(' results/metrics.csv not found - please run 01_EDA_and_Modeling.ipynb first.')
```

```python
path = 'results/plots/eda_distribution.png'
if os.path.exists(path):
 img = mpimg.imread(path)
 plt.figure(figsize=(12, 6))
 plt.imshow(img); plt.axis('off')
 plt.title('EDA: Trip Duration Distribution', fontsize=13, fontweight='bold')
 plt.tight_layout(); plt.show()
else:
 print(f' Plot not found: {path}')
```
